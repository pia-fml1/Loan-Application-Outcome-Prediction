### Detecting and Treating Outliers

```python
# List of columns to handle outliers
columns_to_handle_outliers = ["credit_amount", "duration", "asnm", "age"]

# Create a copy of the original DataFrame
credit_data_capped = credit_data.copy()

# Loop through the specified columns and handle outliers by clipping
for column in columns_to_handle_outliers:
    # Calculate the IQR (Interquartile Range) for the current column
    Q1 = credit_data[column].quantile(0.25)
    Q3 = credit_data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds to identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Clip outliers to the nearest threshold value
    credit_data_capped[column] = credit_data_capped[column].clip(lower=lower_bound, upper=upper_bound)

    # Print the number of outliers capped for each column
    print(f"Number of outliers capped for {column}: {sum((credit_data[column] < lower_bound) | (credit_data[column] > upper_bound))}")
```

### Model Evaluation

```python
from sklearn.metrics import roc_curve, roc_auc_score
# List of all the model names
model_names = ['Logistic Regression', 'K-Nearest Neighbors', 'Support Vector Machines',
                'Random Forest', 'Gradient Boosting',
                'XGBoost']

# List of corresponding fitted models
fitted_models = [logreg, knn_best, svm_best,  random_forest_best, gradient_boosting_best,
                  xgb_best]
# Dictionary to store ROC curve data and AUC-ROC values
roc_data = {}
auc_roc_values = []

# Calculate ROC curve and AUC-ROC for each model
for model_name, model in zip(model_names, fitted_models):
    # Calculate predicted probabilities for positive class
    if hasattr(model, "predict_proba"):
        predicted_probabilities = model.predict_proba(X_test)[:, 1]
    else:
        predicted_probabilities = model.decision_function(X_test)
        
    # Calculate ROC curve (i.e., false positive rate and true positive rate values)
    fpr, tpr, _ = roc_curve(y_test, predicted_probabilities)
    
    # Calculate AUC-ROC (i.e., area under the curve for each model)
    auc_roc = roc_auc_score(y_test, predicted_probabilities)
    auc_roc_values.append(auc_roc)
    
    # Store data in the dictionary
    roc_data[model_name] = {'fpr': fpr, 'tpr': tpr}

# Plot ROC curves for all models
plt.figure(figsize=(7.5, 7.5))
for model_name in model_names:
    plt.plot(roc_data[model_name]['fpr'], roc_data[model_name]['tpr'], label=model_name)

# Plot the random classifier line (dashed)
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc='lower right')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()

# DataFrame for AUC-ROC values
auc_roc_df = pd.DataFrame({'Model': model_names, 'AUC-ROC': auc_roc_values})

# Sort DataFrame by AUC-ROC in descending order
auc_roc_df = auc_roc_df.sort_values(by='AUC-ROC', ascending=False)

# Display AUC-ROC values
print("AUC-ROC Values (Area Under the Receiver Operating Characteristic curve):")
print(auc_roc_df.to_string(index=False))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create a DataFrame to store accuracy results
evaluation_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Evaluate accuracy for each model
for model_name, model in zip(model_names, fitted_models):
    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Store metrics in the DataFrame
    evaluation_df = evaluation_df.append({'Model': model_name, 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}, ignore_index=True)

# Sort the DataFrame by 'Accuracy' column in descending order
evaluation_df_sorted = evaluation_df.sort_values(by='Precision', ascending=False)

# Print the sorted evaluation table
print(evaluation_df_sorted)
```
